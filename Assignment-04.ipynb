```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1 = pd.read_csv('/Users/anandkumar/Downloads/Student file+Assignment 4_Data Vizualization using Python Libraries/Airbnb Dataset 19.csv')
dataset1.head()
```

```python
# Visual 1- Distribution of Neighbourhood Groups
plt.figure(figsize=(10, 6))
sns.countplot(data = dataset1, x="neighbourhood_group", palette="viridis")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Count")
plt.title("Distribution of Neighbourhood Groups")
plt.xticks(rotation=45)
plt.show()
```

```python
# visual-2 No.of reviews and price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dataset1, x="price", y="number_of_reviews")
plt.xlabel("Price")
plt.ylabel("Number of Reviews")
plt.title("Price vs. Number of Reviews")
plt.show()
```

```python
#visual-3 Price distribution of Each Room type
plt.figure(figsize=(8, 5))
sns.boxplot(x='room_type', y='price', data=dataset1, palette='muted')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.title('Price Distribution for Each Room Type')
plt.show()
```

```python
#Visual - 04 Distribution of price
plt.figure(figsize=(8, 5))
sns.histplot(dataset1['price'])
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.show()
```

```python
#Visual - 05 Relation b/w latitude and longitude
plt.figure(figsize=(8, 5))
sns.scatterplot(x='longitude', y='latitude', data=dataset1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Listings')
plt.show()
```

```python
#Dataset-2
dataset2 = pd.read_csv("/Users/anandkumar/Downloads/Student file+Assignment 4_Data Vizualization using Python Libraries/HRDataset_v14.csv")
dataset2.head()
```

```python
#Visual-01 count of employees in each department

plt.figure(figsize=(10, 6))
sns.countplot(x='Department', data=dataset2)
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Number of Employees in Each Department')
plt.xticks(rotation=45)
plt.show()
```

```python
#Visual-02 Distribution of employee salaries
plt.figure(figsize=(8, 5))
sns.histplot(dataset2['Salary'], color='skyblue')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Employee Salaries')
plt.show()
```

```python
#Visual-03 Employee satisfaction score
plt.figure(figsize=(8, 5))
sns.boxplot(x='EmpSatisfaction', data=dataset2)
plt.xlabel('Employee Satisfaction Score')
plt.title('Distribution of Employee Satisfaction Scores')
plt.show()
```

```python
#visual-04 Count of employees in performance score
plt.figure(figsize=(8, 5))
sns.countplot(x='PerformanceScore', data=dataset2)
plt.xlabel('Performance Score')
plt.ylabel('Count')
plt.title('Number of Employees in Each Performance Score Category')
plt.show()
```

```python
#Visual05 - count of employees hired from different recruitment sources
plt.figure(figsize=(10, 6))
sns.countplot(x='RecruitmentSource', data=dataset2)
plt.xlabel('Recruitment Source')
plt.ylabel('Count')
plt.title('Number of Employees Hired from Each Recruitment Source')
plt.xticks(rotation=90)
plt.show()
```
